```python
import torch
from diffusers import DiffusionPipeline, DPMSolverMultistepScheduler
from diffusers.utils import export_to_video

pipe = DiffusionPipeline.from_pretrained("damo-vilab/text-to-video-ms-1.7b", torch_dtype=torch.float16, variant="fp16")   #loading the pretrained model
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe.enable_model_cpu_offload()
```

```python
prompt = input()
video_frames = pipe(prompt, num_inference_steps=25, num_frames=200).frames    #these parameters ==> 25 second video
video_path = export_to_video(video_frames)
```

```python
video_path
```

```python
import os
import shutil
folderName = "Generated_Videos"
if not os.path.exists(folderName):
    os.mkdir(folderName)

file_name = os.path.basename(video_path)
destination_path = os.path.join(folderName, file_name)
shutil.move(video_path, destination_path)       #moving the gerated video to the new folder
```
